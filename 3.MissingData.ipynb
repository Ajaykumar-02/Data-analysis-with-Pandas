```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() #finding missing Values
```

```python
df.isna().sum() # for counting total Na values
```

```python
df.isna().any() # any means every rows missing values ?
```

Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3) # thresh means minimun 3 values in data atlest
```

Filling the missing Data

```python
df
```

```python
df.fillna(0) #fillna value = 0
```

```python
val={"A":0,"B":100,"C":200,"D":400} #Values define
df.fillna(value=val)
```

```python
df
```

```python
df.fillna(df.mean()) #avg
```
